<h1 style="font-size:2rem;color:darkblue;"> Exploratory Data Analysis </h1>

In our data set, each student has 5 distinct properties: Year, Major, University, Time, Order. The <b> main goal </b> of this data analysis is to identify how given the 4 prior distinct properties (Y/M/U/T), one can predict the last property (O). 

<h2 style="color:darkgreen;"> Visualization and Sorting </h2>
<p>In order to determine what other impactful business use cases could be formed with the data, students are first categorized into a bar graph via each property.</p>

The visualization will mainly be focused on <i> category </i> x Students. Although further deductions can be made by comparing each category with every other category, once again for the sake of brevity, this data visualization will only be focusing on <i> category </i> x Students. In addition, there will be a primary focus on properties that have a strong correlation with the observation's interest (what it means for the food truck).

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load the CSV data
data = pd.read_csv('Xtern_TrainData.csv')
```

<h3><b>Year </b> </h3>
By looking at the population per year, it allows one to analyze which years of students are more likely to purchase from a food truck (participate in food commerce), and in return, it reflects what year of students have a better financial background (being able to purchase food rather than cook). 

```python
#Categorize the data
column_name = 'Year'
data[column_name].fillna('Unknown', inplace=True)
category_counts = data[column_name].value_counts()

# Create a bar chart
plt.figure(figsize=(5, 5))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Year')
plt.ylabel('# of Students')
plt.title('Students sorted by Year')
plt.xticks(range(len(category_counts)), category_counts.index, rotation=90)
plt.show()
```

<b> Result: </b> Majority of students purchasing food at the food truck are Sophmores and Juniors. It can be inferred that the workload of those students allow for more free time as compared to Freshmen or Seniors. Another observation is that the Sophmores/Juniors of that year prefer to eat out more, or have the financial means to do so. The food truck can cater heavily towards Sophmores/Juniors through rewards programs or exclusive discounts in order to maximize sales from the overwhelming majority.

<h3> <b> Major</b></h3> 
By looking at the Majors of students, it can help show the workload of these majors and the amount of free time a student has, however from a quick overview, the data shows zero correlation, with the overall distribution of students showing varying degrees of randomness in regards to the food truck.

<h3> <b> University</b></h3> 
By looking at the population per University, it allows one to analyze which locations have students that are more inclined to purchase food from a food truck due to accessibility, and can reflect the quality of a Universities food courts or environment (restaurants, food chains, etc). 

```python
#Categorize the data
column_name = 'University'
data[column_name].fillna('Unknown', inplace=True)
data[column_name] = data[column_name].replace('Indiana University-Purdue University Indianapolis (IUPUI)', 'IUPUI')
category_counts = data[column_name].value_counts()

# Create a bar chart
plt.figure(figsize=(5, 5))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('University')
plt.ylabel('# of Students')
plt.title('Students sorted by Uni')
plt.xticks(range(len(category_counts)), category_counts.index, rotation=90)
plt.show()
```

<b> Result: </b> Majority of students purchasing food at the food truck are from Butler University, Indiana State University, Ball State University, and IUPUI. It can be inferred that these locations feature students that have a wider range of accessibility to the food truck, whether it is distance, food court quality, restaurant quality, or other region based properties.

<h3> <b> Time</b></h3> 
Time is a very important factor for students, and therefore it is important for this data visualization to take a closer look at it as well. Being arguably the most valuable currency for students, a quick visualization of time can represent when most students eat their breakfast/lunch/dinner, and therefore reflect rush hour. For food trucks, being able to know what time most students dine can be the difference between wasting gasoline and gaining a quick profit.

```python
#Re-load the CSV data without interpreting columns as strings for Time
datat = pd.read_csv('Xtern_TrainData.csv')

#Categorize the data
column_name = 'Time'
datat[column_name].fillna('Unknown', inplace=True)
encoder = LabelEncoder()
datat[column_name] = encoder.fit_transform(datat[column_name])
category_counts = datat[column_name].value_counts().sort_index()

# Create a bar chart
plt.figure(figsize=(5, 5))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Time')
plt.ylabel('# of Students')
plt.title('Students sorted by Time')
plt.xticks(category_counts.index, encoder.inverse_transform(category_counts.index))
plt.show()
```

<b> Result: </b> Majority of students purchase food around lunchtime (11-14), with the lunch rush specifically being around (12-13). This data allows food trucks to meal prep beforehand in order to go through as many customers as possible. In addition, it can help reduce opertaional costs for times where there is little to no consumer activity.

<h3><b>Order</b></h3>
By looking at the order per year, it allows food services such as the food truck in question to gage the amount of stock they should order, in order to not only optimize cost but reduce waste as well. It also allows one to raise the price on a food item in high demand in order to either generate a larger profit, usher customers into ordering other menu items, or both. 

```python
#Categorize the data
column_name = 'Order'
data[column_name].fillna('Unknown', inplace=True)
data[column_name] = data[column_name].replace('Ultimate Grilled Cheese Sandwich (with bacon and tomato)', 'UGCS w/ b&t')
data[column_name] = data[column_name].replace('Indiana Corn on the Cob (brushed with garlic butter)', 'ICC w/ gb')
category_counts = data[column_name].value_counts()

# Create a bar chart
plt.figure(figsize=(5, 5))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Order')
plt.ylabel('# of Students')
plt.title('Students sorted by Order')
plt.xticks(range(len(category_counts)), category_counts.index, rotation=90)
plt.show()
```

<b> Result: </b> The order distribution is overall evenly distributed, showing that every item is in relative good demand. The food truck can determine whether or not they want to modify their prices in order to drive customers towards one particular item, or keep the current demand the same.

<h1 style="font-size:2rem;color:darkblue;"> Prediction Model </h1>

<h2 style="color:darkgreen;"> 1. Model Selection </h2>

The model used for order prediction will be <b> Random Forest </b> classifier via Regression, as it is the best suited model for predicting categories due to the nature of the task, which is classification. The training model will be the given <h>Xtern_TrainData.csv</h> data set, and we will be testing using an example student input via the SKLearn .predict() method. We will first load our data using a pandas DataFrame.

<h2 style="color:darkgreen;"> 2. Loading Data </h2>

To properly handle data preparation and loading, the model needs to make sure that it handles missing data, encodes categorical data, and splits the data into training/testing sets. For our purposes, the testing set will consist of 20% of our data.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle

# Load the CSV data
data = pd.read_csv('Xtern_TrainData.csv')

# Handle missing data
data['University'].fillna('Unknown', inplace=True)
data['Year'].fillna('Unknown', inplace=True)
data['Major'].fillna('Unknown', inplace=True)
data['Time'].fillna('Unknown', inplace=True)

# Split selection and target category
x = data[['University', 'Year', 'Major', 'Time']]
y = data['Order']

# Label encoding for categorical features
label_encoders = {}
for col in x.columns:
    label_encoders[col] = LabelEncoder()
    x.loc[:, col] = label_encoders[col].fit_transform(x.loc[:, col])

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

<h2 style="color:darkgreen;"> 2. Model Training and Pickle </h2>

The classifier used to train the data is Random Forest. We can first train our data and then save the trained data via the pickle.dump function. For the sake of this model, the data is only trained once. Therefore the loaded model is brought back and used to predict data using the .predict() function.

```python
# Choose and train a classifier (e.g., Random Forest)
classifier = RandomForestClassifier()
classifier.fit(x_train, y_train)

# Model Training
y_pred = classifier.predict(x_test)

# Model pickle and loading.
with open('XTern_TrainedData.pkl', 'wb') as model_file:
    pickle.dump(classifier, model_file)

with open('XTern_TrainedData.pkl', 'rb') as model_file:
    loaded_model = pickle.load(model_file)
```

<h2 style="color:darkgreen;"> 3. Training and Testing </h2>

After our data is trained, we can test the model prediction accuracy on various test students, and compare them to preexisting students which are similar.

```python
# Example Customer 1
example_student = pd.DataFrame({'University': ['University of Notre Dame'], 'Year': ['Year 2'], 'Major': ['Business Administration'], 'Time': ['15']})
print(example_student)

for col in example_student.columns:
    example_student[col] = label_encoders[col].transform(example_student[col])

predicted_order = loaded_model.predict(example_student)
print("Predicted Order for New Student:", predicted_order)
```

This result lines up with a preeisting customer, which matched the same description except for year, and shared the same order.

```python
# Example Customer 2
example_student = pd.DataFrame({'University': ['Indiana State University'], 'Year': ['Year 3'], 'Major': ['Business Administration'], 'Time': ['13']})
print(example_student)

for col in example_student.columns:
    example_student[col] = label_encoders[col].transform(example_student[col])

predicted_order = loaded_model.predict(example_student)

print("Predicted Order for New Student:", predicted_order)
```

This result lines up with sample points which share the same University, Year, and Time, but differ in Order. However the order lines up with the most common Order for Business Administration majors.

<b> Accuracy </b> We can check for general accuracy using SKLearn's accuracy_score function. Although this checks for exact results, it can still be useful to gage the current usefulness of our model.

```python
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
```

As we can see, the model accuracy is roughly 65.2%. So for every 10 customers, 4 discounts need to be given.
